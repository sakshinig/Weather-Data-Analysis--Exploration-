```python
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
import numpy as np
import pandas as pd
```

```python
df = pd.read_csv('05-2019.csv')
df1 = pd.read_csv('WeatherData/06-2019.csv')
df2 = pd.read_csv('WeatherData/07-2019.csv')
df3 = pd.read_csv('WeatherData/08-2019.csv')
df4 = pd.read_csv('WeatherData/09-2019.csv')
df5 = pd.read_csv('WeatherData/10-2019.csv')
df6 = pd.read_csv('WeatherData/11-2019.csv')
df7 = pd.read_csv('WeatherData/12-2019.csv')

data = pd.concat([df, df1, df2, df3, df4, df5, df6, df7])
```
